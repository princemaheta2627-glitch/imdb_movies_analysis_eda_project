```python
votes_by_year = df.groupby('Year')['Votes'].sum()
votes_by_year.plot(kind='bar', figsize=(10,5), title="Votes by Year")
plt.show()
#analysis by voting
```

### 🔍 Explanation
The groupby() function is used to group data based on specific categories. It allows aggregation such as average, sum, or count, helping uncover patterns across different groups.

### 📊 Insight
This output helps in understanding the data better. Based on this result, we can interpret patterns, trends, or anomalies that are useful for further analysis.

```python
revenue_by_year = df.groupby('Year')['Revenue (Millions)'].sum()
revenue_by_year.plot(kind='line', marker='o', figsize=(10,5), title="Revenue by Year")
plt.show()
#revenue trends 
```

### 🔍 Explanation
The groupby() function is used to group data based on specific categories. It allows aggregation such as average, sum, or count, helping uncover patterns across different groups.

### 📊 Insight
This output helps in understanding the data better. Based on this result, we can interpret patterns, trends, or anomalies that are useful for further analysis.

```python
director_ratings = df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(10)
print("Top Directors by Average Rating:")
print(director_ratings)
#dircector performance
```

### 🔍 Explanation
The head() function displays the first 5 rows of the dataset. This helps in understanding how the data looks, identifying column names, and getting an initial sense of the values.

### 📊 Insight

```python
movies_per_year = df['Year'].value_counts().sort_index()
movies_per_year.plot(kind='bar', figsize=(10,5), title="Movies Produced per Year")
plt.show()
#movie production as per year
```

### 🔍 Explanation
value_counts() counts the occurrences of each unique value in a column. This helps understand the distribution of categorical data.

### 📊 Insight
This output helps in understanding the data better. Based on this result, we can interpret patterns, trends, or anomalies that are useful for further analysis.

```python
def classify_movie(rating):
    if rating >= 8:
        return "Excellent"
    elif rating >= 6:
        return "Good"
    else:
        return "Average"

df['Category'] = df['Rating'].apply(classify_movie)
print(df[['Title','Rating','Category']].head())
#classification of movies
```

### 🔍 Explanation
The head() function displays the first 5 rows of the dataset. This helps in understanding how the data looks, identifying column names, and getting an initial sense of the values.

### 📊 Insight
This output helps in understanding the data better. Based on this result, we can interpret patterns, trends, or anomalies that are useful for further analysis.

```python
genre_revenue = df.groupby('Genre')['Revenue (Millions)'].mean().sort_values(ascending=False)
print(genre_revenue)
#genre wise revenue
```

### 🔍 Explanation
